We check for missing values in each column to decide how to handle them.


```python
df.isnull().sum()
```

# Step 4: Fill Missing Values
- company → filled with `"None"`  
- agent → filled with `"Unknown"`  
- country → filled with `"Unknown"`  
- children → filled with `0`


```python
df["company"] = df["company"].fillna("None")
df["agent"] = df["agent"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
df["children"] = df["children"].fillna(0)
```

# Step 5: Duplicates
We check for duplicate rows and remove them.


```python
print("Duplicate rows:", df.duplicated().sum())
df = df.drop_duplicates()
```

# Step 6: Outliers in ADR
ADR (Average Daily Rate) should not be negative and extremely high values are unrealistic.  
We set negatives to NaN and cap ADR at 1000.


```python
print("ADR min:", df["adr"].min(), "max:", df["adr"].max())
df.loc[df["adr"] <= 0, "adr"] = None   # set negative to missing
df["adr"] = df["adr"].clip(upper=1000) # cap extreme highs
```

# Step 7: Invalid Rows
We remove bookings where total guests = 0 (invalid bookings).


```python
df["total_guests"] = df["adults"] + df["children"] + df["babies"]
print("Zero-guest rows:", (df["total_guests"]==0).sum())
df = df[df["total_guests"] > 0]
```

# Step 8: Drop Leakage Columns
We drop `reservation_status` and `reservation_status_date` because they reveal the booking outcome.


```python
df = df.drop(columns=["reservation_status", "reservation_status_date"])
```

# Step 9: Save Cleaned Data
We save the cleaned dataset as `hotel_bookings_cleaned.csv`.


```python
df.to_csv("hotel_bookings_cleaned.csv", index=False)
```
